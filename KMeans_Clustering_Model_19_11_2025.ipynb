```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

image.png

```python
df = pd.read_csv('/content/income.csv')
df
```

```python
df.isna().sum()
```

```python
plt.scatter(df['Age'],df['Income($)'])
plt.xlabel('Age')
plt.ylabel('Income($)')
```

```python
km = KMeans(n_clusters=3)
```

```python
y_pred = km.fit_predict(df[['Age','Income($)']])
y_pred
```

```python
df['Cluster'] = y_pred
df
```

```python
km.cluster_centers_
```

```python
df0 = df[df.Cluster==0]
df1 = df[df.Cluster==1]
df2 = df[df.Cluster==2]
```

```python
plt.scatter(df0['Age'],df0['Income($)'],color='green')
plt.scatter(df1['Age'],df1['Income($)'],color='red')
plt.scatter(df2['Age'],df2['Income($)'],color='black')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.legend()
plt.show()
```

```python
#We observed a scaling issue

scaler = MinMaxScaler()


scaler.fit(df[['Income($)']])
df['Income($)'] = scaler.transform(df[['Income($)']])


scaler.fit(df[['Age']])
df['Age'] = scaler.transform(df[['Age']])
```

```python
km1 = KMeans(n_clusters=3)

y_pred1 = km1.fit_predict(df[['Age','Income($)']])

df['Cluster1'] = y_pred1
df
```

```python
#New clusters
df0 = df[df.Cluster1==0]
df1 = df[df.Cluster1==1]
df2 = df[df.Cluster1==2]
```

```python
plt.scatter(df0['Age'],df0['Income($)'],color='green')
plt.scatter(df1['Age'],df1['Income($)'],color='red')
plt.scatter(df2['Age'],df2['Income($)'],color='black')
plt.scatter(km1.cluster_centers_[:,0],km1.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.xlabel('Age')
plt.ylabel('Income')
plt.legend()
plt.show()
```
